–†–µ–∞–ª–∏–∑–æ–≤–∞—Ç—å –∞–ª–≥–æ—Ä–∏—Ç–º –ë—Ä–∞—É—ç—Ä–∞ –¥–ª—è –≤—ã—á–∏—Å–ª–µ–Ω–∏—è –ø—Ä–∏–±–ª–∏–∂—ë–Ω–Ω—ã—Ö –∞–¥–¥–∏—Ç–∏–≤–Ω—ã—Ö —Ü–µ–ø–æ—á–µ–∫ –¥–ª—è —Ä–∞–∑–ª–∏—á–Ω—ã—Ö —á–∏—Å–µ–ª –ø—Ä–∏ –≤–∞—Ä—å–∏—Ä–æ–≤–∞–Ω–∏–∏ –ø–∞—Ä–∞–º–µ—Ç—Ä–∞ ùëò, —Å–æ–ø–æ—Å—Ç–∞–≤–∏—Ç—å –¥–ª–∏–Ω—ã –ø–æ–ª—É—á–µ–Ω–Ω—ã—Ö –∞–¥–¥–∏—Ç–∏–≤–Ω—ã—Ö —Ü–µ–ø–æ—á–µ–∫ —Å –º–∏–Ω–∏–º–∞–ª—å–Ω–æ–π –∞–¥–¥–∏—Ç–∏–≤–Ω–æ–π —Ü–µ–ø–æ—á–∫–æ–π –¥–ª—è –∑–∞–¥–∞–Ω–Ω–æ–≥–æ —á–∏—Å–ª–∞. –°–¥–µ–ª–∞—Ç—å –≤—ã–≤–æ–¥—ã.

```python
def brauer():
    n=int(input())
    k=int(input())
    d=2^k
    r=[]
    leftovers=[]
    subsqence=[]
    buff=0
    coefficients=0
    while(n>d):
        r.append(n%d)
        n=n//d
        leftovers.append(n)
    
    for i in range(1, d):
        subsqence.append(i)
        
        
    for i in range(1, len(r)+1):
        coefficients=2
        buff=leftovers[-i]
        while coefficients <= d:
            subsqence.append(leftovers[-i]*coefficients)
            coefficients = coefficients*2
        subsqence.append(buff*coefficients/2+r[-i])
        
    print(sorted(subsqence))
        
brauer()
```

```python
from math import floor, log
from datetime import datetime


def get_chain(vec):  
    line = []
    line.append(1)
    for i in range(len(vec)):
        line.append(line[i] + line[vec[i] - 1])
    return line


def dect(vec): 
    for i in vec:
        if i > 1:
            return False
    return True

def dec(vec):
    _len = len(vec)
    for i in range(_len):
        if vec[_len - i - 1] > 1:
            vec[_len - i - 1] -= 1
            for j in range(_len - i, _len):
                vec[j] = j + 1
            break





def vector_index(n):
    
    BINAR = format(n, 'b')
    LAMBDA = len(BINAR) - 1
    NU = BINAR.count('1')
    MAX_l = LAMBDA + NU - 1
    MIN_l = LAMBDA + floor(log(n, 2))
    tries = [1, 1]

    for m in range(MAX_l, MIN_l):
        q = int(m / 2)

        vector_part = [0] * q
        for i in range(q):
            vector_part[i] = 1 + i

        vector_fix = vector_part

        while True:  

            change_part = [0] * (m - q)
            for i in range(m - q):
                change_part[i] = q + 1 + i

            change = change_part

            bounds = [0] * 2

            a = []
            a.append(1)
            for i in range(len(vector_fix)):
                a.append(a[i] + a[vector_fix[i] - 1])
            a_1 = a[len(a) - 1]

            bounds[0] = a_1 + m - q
            bounds[1] = a_1 * 2 ** (m - q)

            if n < bounds[0] or n > bounds[1]:
                if dect(vector_fix) and len(vector_fix) > 1:
                    break
                dec(vector_fix)
                continue

            while True:  

                vector_full = []
                vector_full.append(1)
                for i in range(len(vector_fix + change)):
                    vector_full.append(vector_full[i] + vector_full[(vector_fix + change)[i] - 1])
                vector_full = vector_full[len(vector_full) - 1]

                if vector_full == n:
                    return vector_fix + change

                if dect(change):
                    break
                dec(change)
            if dect(vector_fix):
                break
            dec(vector_fix)


n = int(input())
time_1 = datetime.now()
vova = vector_index(n)

print("chain : ", get_chain(vova))
time_2 = datetime.now()
print(time_2-time_1)
```
